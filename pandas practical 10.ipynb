#fillna()..to fill the missing or NaN values by given input

```python
import pandas as pd
```

```python
df=pd.read_csv('D:\Movies\BOLLYWOOD\Flames\practice.csv')
df
```

```python
df.fillna(0)
```

#to print the different values in different column as per requirement so we use dictionary

```python
df.fillna({'A':'utkarsh','C':'Lapi'})
```

#method()..to fill forward or backward values in place of NaN

```python
df.fillna(method='ffill')
```

```python
df.fillna(method='pad')
```

```python
df.fillna(method='bfill')
```

#axis()...it must with method() otherwise gives error

```python
df.fillna(axis=0,method='ffill')
```

```python
df.fillna(axis=1,method='ffill')
```

#limit()..max limit NaN change by its value or method in particular column

```python
df
```

```python
df.fillna(0,limit=1) #by value
```

```python
df.fillna(0,limit=2)
```

```python
df
```

```python
df.fillna(method='ffill',limit=1)
```

```python
df.fillna(method='bfill',limit=1)
```

#inplacce()..exchane NaN value with given input and does not create new dataframe modify in current dataframe

```python
df.fillna(5,inplace=True)
df
```

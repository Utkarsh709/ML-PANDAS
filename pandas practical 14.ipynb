#slicing

```python
df.loc[0:3]
```

```python
df.loc[0:3,'study']
```

```python
df.loc[0:]
```

#obtaining by boolean list

```python
df.loc[[False,True,True,False,True]]
```

#conditional operator

```python
#if want the values of column class which contain less than 11 values
df.loc[df['class']<11]
```

```python
#i want above condition and want to print only values of column media 
df.loc[df['class']<11,['media']]
```

#dropna() method it removes row if it contain NaN values by default axis=0


```python
import pandas as pd
```

```python
df=pd.read_csv('D:\Movies\BOLLYWOOD\Flames\practice.csv')
df
```

```python
df.dropna()
```

```python
df.dropna(axis=0)
```

```python
df.dropna(axis=1)
```

#df.dropna(how='any')...similar like df.dropna()

```python
df.dropna(how='any')
```

```python
df.dropna(how='any',axis=1)  #df.dropna(axis=1)
```

#df.dropna(how='all')...if rows contain all null values then it can deop that row

```python
df=pd.read_csv('D:\Movies\BOLLYWOOD\Flames\practice.csv')
df
```

```python
df.dropna(how='all')
```

#thresh()..it print bydefault rows having value at least one not a null value 

```python
df.dropna(thresh=1)
#having atlest one not null value
```

```python
df.dropna(thresh=4)
```

#subset()...drop a column having having NaN value

```python
df
```

```python
df.dropna(subset=["C"])
```

#Series#

```python
import pandas as pd
```

```python
pd.__version__
```

```python
list1=[1,2,3,True,"Utkarsh"]
print(list1)
```

```python
series1=pd.Series(list1)
print(series1)
```

```python
type(series1)
```

```python
series2=pd.Series([5,6,7,"UVK",False])
print(series2)
```

```python
empty_s=pd.Series([])
print(empty_s)
```

#To change the index value

```python
series3=pd.Series([1,2,3,4],index=['a','b','c','d'])
print(series3)
```

#To change the datatype

```python
series3=pd.Series([1,2,3,4],index=['a','b','c','d'],dtype=float)
print(series3)
```

```python
series3=pd.Series([1,2,3,4],index=['a','b','c','d'],dtype=bool)
print(series3)
```

#To give the name

```python
series3=pd.Series([1,2,3,4],index=['a','b','c','d'],dtype=float,name="Hours")
print(series3)
```

#single Scalar Values

```python
scalar_s=pd.Series(0.5)
print(scalar_s)
```

```python
scalar_s=pd.Series(0.5,index=[1,2,3,4,5,6])
print(scalar_s)
```

#Creating series with help of dictionary...key become index and values become object

```python
dict_s=pd.Series({'a':1,'b':2})
print(dict_s)
```

#Slicing

```python
s4=pd.Series([4,5,6,7,8,9])
print(s4)
```

```python
s4[0]
```

```python
s4[0:4]
```


#To find max and min value

```python
max(s4)
```

```python
min(s4)
```

#Conditional Operator

```python
s4[s4>6]
```

#Mathematical Opeartion

```python
s5=pd.Series([1,2,3,4])
s6=pd.Series([5,6,7,8])
print(s5)
print(s6)
print("Addition")
print(s5+s6)
print("Subtraction")
print(s5-s6)
print("Divison")
print(s5/s6)
print("Multiplication")
print(s5*s6)
```

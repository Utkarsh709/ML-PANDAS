```python
import pandas as pd
```

```python
df1=pd.DataFrame({
       'A':[1,2,3],
        'B':[10,20,30]
})

df2=pd.DataFrame({
        'C':[4,5,6],
        'D':[40,50,60]
    
})
display(df1,df2)
```

```python
df1.join(df2)
```

```python
df3=pd.DataFrame({
       'A':[1,2,3],
        'B':[10,20,30]},
    index=['a','b','c']
)
df3
```

```python
df3.join(df2)
```

```python
df4=pd.DataFrame({
       'C':[4,5],
        'D':[40,50]},
    index=['a','b']
)
df4
```

```python
df3.join(df4)
```

```python
df3.join(df4,how='outer')
```

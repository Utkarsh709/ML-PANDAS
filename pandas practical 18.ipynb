```python
import pandas as pd
```

```python
df1=pd.DataFrame({'id':[1,2,3,4],
                  'name':['A','B','C','D'],
                  'class':[5,6,7,8]

    
})
df1
```

```python
df2=pd.DataFrame({'id':[3,4],
                  'name':['C','D'],
                  'class':[7,8]

    
})
df2
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2],axis=1)
```

```python
pd.concat([df1,df2],axis=1,join='inner') #intersection
```

```python
pd.concat([df1,df2],axis=1,join='outer')   #union
```

```python
pd.concat([df1,df2],axis=1,ignore_index=[df1.index])
```

```python
pd.concat([df1,df2],axis=1,ignore_index=[df2.index])
```

```python
pd.concat([df1,df2],keys=['df1','df2'])
```

```python
pd.concat([df1,df2],keys=['df1','df2'],axis=1)
```

#pandas merging() function...

```python
import pandas as pd
```

```python
df1=pd.DataFrame({'id':[1,2,3,4],
                'class':[9,10,11,12]})
df1
```

```python
df2=pd.DataFrame({'id':[1,2,3,4],
                 'name':['A','B','C','D']})
df2
```

```python
pd.merge(df1,df2,on='id')
```

```python
pd.merge(df2,df1,on='id')
```

#if id colummn is not same

```python
df3=pd.DataFrame({'id':[1,2,3,5],
                 'name':['A','B','C','D']})
df3
```

```python
pd.merge(df1,df3,on='id')
```

#how 

```python
pd.merge(df1,df3,on='id',how='inner')   #by default it take how=inner..means intersection
```

```python
pd.merge(df1,df3,on='id',how='left') #how=left give prioritize left one in this priority =df1
```

```python
pd.merge(df1,df3,on='id',how='right') #how=right give prioritize right one in this priority =df3
```

```python
pd.merge(df1,df3,on='id',how='outer') #how=outer means union
```

#indicator paramter to know the information of each row

```python
pd.merge(df1,df3,on='id',how='outer',indicator='True')  #by default indicator take value False
```

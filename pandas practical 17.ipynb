#concat()

```python
import pandas as pd
```

#series...

```python
sr1=pd.Series([0,1,2])
sr1
```

```python
sr2=pd.Series([3,4,5])
sr2
```

```python
pd.concat([sr1,sr2])
```

```python
sr3=pd.Series([3,4,5,6,7])
sr3
```

```python
pd.concat([sr1,sr3])
```

#DataFrame..

```python
df1=pd.DataFrame({'id':[1,2,3,4],
                  'name':['A','B','C','D'],
                  'class':[5,6,7,8]

    
})
df1
```

```python
df2=pd.DataFrame({'id':[5,6,7,8],
                  'name':['E','F','G','H'],
                  'class':[9,10,11,12]

    
})
df2
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2],axis=1)
```

```python
df3=pd.DataFrame({'id':[5,6,7,8,9],
                  'name':['E','F','G','H','I'],
                  'class':[9,10,11,12,13]

    
})
df3
```

```python
pd.concat([df1,df3],axis=1)
```

```python
pd.concat([df1,df3],axis=0)
```

#to make the index systematic

```python
pd.concat([df1,df3],axis=0,ignore_index=True)
```

```python
#Dictionary
```

```python
dict1={'ID':[1,2,3,4]}
print(dict1)
```

```python
df3=pd.DataFrame(dict1)
df3
```

```python
dict1={'ID':[1,2,3,4],'SN':[6,7,8,9]}
print(dict1)
df3=pd.DataFrame(dict1)
df3
```

```python
#list of dictionaries
```

```python
list_of_dict=[{'a':1},{'b':2},{'a':4,'b':8,'c':9}]
print(list_of_dict)
df4=pd.DataFrame(list_of_dict)
df4
```

```python
#dictionaries of Series
```

```python
dict_of_ser={'a':pd.Series([1,2,3,4]),'b':pd.Series([6,7])}
print(dict_of_ser)
df5=pd.DataFrame(dict_of_ser)
df5
```

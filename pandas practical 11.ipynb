#replace()..to replace existing values in dataframe with other value

```python
import pandas as pd
```

```python
df=pd.read_csv('D:\\Movies\\BOLLYWOOD\\Flames\\book.csv')
df
```

```python
df.replace(to_replace='health',value='mobile')
```

```python
df.replace('health','mobile')  #df.replace(old,new)
```

#by using list

```python
df.replace([1,2,3,4,5],8)
```

```python
df.replace([1,2,3,4,5],[7,8,9,10,11])
```

#to change the value of particular column

```python
df
```

```python
df.replace({'industry':'health','profit':30},'good')
```

#to change the all string with given input by finding pattern using regex

```python
df.replace('[A-Za-z]',0)
```

```python
#using regex()..to detect pattern recognition
df.replace('[A-Za-z]',0,regex=True)
```

#regex for paticular column so use dictionary

```python
df.replace({'industry':'[A-Za-z]'},'aman',regex=True)
```

```python
df.replace({'industry':'[A-Za-z]'},5,regex=True)
```

#method()..fills with forward values

```python
df
```

```python
df.replace('security',method='ffill')
```

```python
df.replace('security',method='bfill')
```

```python
df.replace(30,method='ffill',limit=1)
```

#inplace()..it does not create new datafrane it modifys in old dataframe...replace(old,new,inplace=True)

```python
df.replace('health','jio',inplace=True)
df
```
